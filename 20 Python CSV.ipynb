```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import itertools
import csv
```

```python
files = ['C:/Users/Elite/رفع الى GitHub/Restaurant.csv'] # polarity,text,restaurant_id,user_id
```

```python
def prepareDatasets(files):
    sentences = []

    for file_name in files:
        file = open(file_name, "rt", encoding="utf-8-sig")
        reviews = csv.reader(file)
        
        next(reviews, None)   # لتجاوز أول صف header
        
        for polarity, text, restaurant_id, user_id in reviews:
            sentences.append([text, polarity])

        file.close()

    df = DataFrame(sentences, columns=['text', 'polarity'])
    return df
```

```python
reviews = prepareDatasets(files)
```
