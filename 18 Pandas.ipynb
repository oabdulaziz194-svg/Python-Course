```python
left = DataFrame({'group': ['a', 'a', 'a', 'b','b', 'b', 'c', 'c','c'],
                 'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

left
```

```python
right = DataFrame({'label': ['a','b','c'],
                     'value': ['alpha','beta','charlie']})
right
```

```python
inner_joined = pd.merge(left, right, how='inner', left_on='group', right_on='label')
inner_joined
```
